```python
import pandas as pd

# Leitura da base de Dados
df = pd.read_csv("Base_Uber.csv")
df.head(2)
```

```python
# Substituir valores nulos 
df["Reason for cancelling by Customer"] = df["Reason for cancelling by Customer"].fillna("There was no cancellation")
df["Cancelled Rides by Driver"] = df["Cancelled Rides by Driver"].fillna("There was no cancellation")
df["Driver Cancellation Reason"] = df["Driver Cancellation Reason"].fillna("There was no cancellation")
df["Incomplete Rides"] = df["Incomplete Rides"].fillna("0")
df["Incomplete Rides Reason"] = df["Incomplete Rides Reason"].fillna("Complete Race")
df["Booking Value"] = df["Booking Value"].fillna("0")
df["Ride Distance"] = df["Ride Distance"].fillna("0")
df["Customer Rating"] = df["Customer Rating"].fillna("0")
df["Avg VTAT"] = df["Avg VTAT"].fillna("0")
df["Avg CTAT"] = df["Avg CTAT"].fillna("0")

# Tirar as Aspas
df["Booking ID"] = df["Booking ID"].str.replace('"','').str.strip()
df["Customer ID"] = df["Customer ID"].str.replace('"','').str.strip()
```

```python
linhas, colunas = df.shape
print(f"A Tabela Possui {linhas} Linhas e {colunas} Colunas.")
```
